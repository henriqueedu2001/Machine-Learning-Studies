Regress√£o Linear

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#leitura do dataset
df = pd.read_csv('Iris.csv')
df.head()
```

```python
# vamos selecionar duas features que parecem ter uma correla√ß√£o linear
X = df['PetalLengthCm']
y = df['PetalWidthCm']

# exibe plot scatter das duas features
plt.title("PetalLength (Cm) vs PetalWidth (Cm)")
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")

plt.scatter(X, y)
plt.show()
```

```python
# divide o dataset em 75% para treino e 25% para teste 
train_fraction = 0.75
X_train, x_test, Y_train, y_test = train_test_split(X, y, train_size=train_fraction)
```

```python
# exibi√ß√£o do dataset ap√≥s divis√£o
plt.scatter(X_train, Y_train, label='treino', c='blue')
plt.scatter(x_test, y_test, label='teste', c='red')

plt.title("Divis√£o 75% Treino e 25% Teste")
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.legend()
plt.show()
```

```python
# treinamento do regressor
linear_regressor = LinearRegression()
linear_regressor.fit(X_train.values.reshape(-1, 1), Y_train)
```

```python
# coeficientes obtidos pelo regressor
print('coeficiente linear: ', linear_regressor.intercept_)
print('coeficiente angular: ', linear_regressor.coef_[0])
```

```python
# visualiza√ß√£o da regress√£o
predictionLine = linear_regressor.predict(np.arange(0, 10).reshape(-1, 1))

plt.scatter(X_train, Y_train, label='treino', c='blue')
plt.scatter(x_test, y_test, label='teste', c='red')
plt.plot(predictionLine, label='regress√£o', c='purple')

plt.title("Regress√£o linear sobre as features petallenght e PetalWidth")
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.legend()
plt.show()
```

```python
# c√°lculo das pontu√ß√µes para os datasets de treino e de teste
train_score = linear_regressor.score(X_train, Y_train)
test_score = linear_regressor.score(x_test.values.reshape(-1, 1), y_test)

print('pontua√ß√£o sobre dataset de treino', train_score)
print('pontua√ß√£o sobre dataset de teste', test_score)
```

```python
petal_lengths = np.array([3, 5, 6])
petal_widths = linear_regressor.predict(petal_lengths.reshape(-1, 1))
print(petal_lengths)
print(petal_widths)
```

```python
# predi√ß√£o

plt.scatter(X_train, Y_train, label='treino', c='blue')
plt.scatter(x_test, y_test, label='teste', c='red')
plt.scatter(petal_lengths, petal_widths, label='predi√ß√£o', c='green')

plt.title("Predi√ß√£o")
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.legend()
plt.show()
```
